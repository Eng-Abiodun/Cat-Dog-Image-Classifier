```python
import numpy as np
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import os
import tensorflow as tf
from  tensorflow import keras
```

```python
train_dir= r'C:\Users\ROYALTY\ImageClassification\dog_cat_classification2\data\train'
validation_dir=r'C:\Users\ROYALTY\ImageClassification\dog_cat_classification2\data\validation'
test_dir=r'C:\Users\ROYALTY\ImageClassification\dog_cat_classification2\data\test'
#path to where the images are located
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
#generating the batches of tensor image data
train_datagen=ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,) #we applied all these to create a dummy data from our original data, to enhance our model more

test_datagen=ImageDataGenerator(rescale=1./255)

train_generator= train_datagen.flow_from_directory(
        train_dir,
        target_size=(150,150),
        batch_size=32,
        class_mode='binary')

validation_generator= test_datagen.flow_from_directory(
        validation_dir,
        target_size=(150,150),
        batch_size=32,
        class_mode='binary')
```

```python
batch=train_generator.next()
```

```python
fig, ax= plt.subplots(ncols=4,figsize=(20,20))
for idx, img in enumerate(batch[0][:4]):
    ax[idx].imshow(img)
    ax[idx].title.set_text(batch[1][idx])
#this will print out 4 images randomly from 2 classes, to show whether cat or dog is representing either 1 or 0
```

### Dog is 1
### Cat is 0

```python
from tensorflow.keras import layers
from tensorflow.keras import models
```

```python
model= models.Sequential()
model.add(layers.Conv2D(32,(3,3),activation='relu',input_shape=(150,150,3)))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Conv2D(64,(3,3),activation='relu',input_shape=(150,150,3)))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Conv2D(128,(3,3),activation='relu',input_shape=(150,150,3)))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Conv2D(128,(3,3),activation='relu',input_shape=(150,150,3)))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Flatten())
model.add(layers.Dropout(0.5))#this layer will deactivate 50% of neurons in each epoch randomly
model.add(layers.Dense(512,activation='relu'))
model.add(layers.Dense(1,activation='sigmoid'))

from tensorflow.keras import optimizers
model.compile(loss='binary_crossentropy',
             optimizer=optimizers.RMSprop(learning_rate=1e-4),
             metrics=['acc'])
```

```python
model.summary()
```

```python
hist = model.fit(
    train_generator,
    epochs=100,
    validation_data=validation_generator
)
#might take hours to run depending on your system RAM
```

```python
import pandas as pd
pd.DataFrame(hist.history).plot(figsize=(8,5))
plt.grid(True)
plt.gca().set_ylim(0,1)
plt.show()
```
